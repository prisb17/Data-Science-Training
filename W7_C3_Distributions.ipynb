# Creating and Writing Data

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Normal distribution

```python
# generate random values knowing men, standard deviation and # of values in the series
series = np.random.normal(10, 2, 50)
plt.hist(series)
plt.show()
```

## Uniform distributions

```python
#np.random.uniform (min = x, max = y, n), where n is the number of values in the data
#series and the min and max parameters set the upper and lower bounds. 
series1=np.random.uniform(10, 20, 50)
plt.hist(series1)
plt.show()
```

## Poisson distributions

```python
# random.poisson (lambda, n), where n is the number of values in the data series and
# lambda sets the lambda value. 
series2=np.random.poisson (3, 50)
plt.hist(series2)
plt.show()
```

## Binomial distributions

```python
# np.random.binomial (n, prob, size), where n is the number of observations you want,
# size is the number of trials per observation, and prob is the probability of success for
# each trial. For example, to estimate the outcome of flipping a single coin 100 times
series3=np.random.binomial(100, 0.5, 1)
plt.hist(series3)
plt.show()

# Or, to flip 1000 coins 10 times each: np.random.binomial(10, 0.5, 1000)
```

```python
from plotnine.data import mtcars
mtcars.info()
mtcars
```

```python
zero_sixty=1/(mtcars['hp']/mtcars['wt'])*440
mtcars['zero to sixty']=zero_sixty
# the row above adds a column to the dataframe
mtcars
```

```python
# Graph scatter plot showing the relationship between the 0-60 time and the quarter-mile time
from matplotlib import *
style.use('ggplot')
plt.plot(mtcars['zero to sixty'],mtcars['qsec'],'.')

# setting axis labels and plot title
plt.xlabel('0-60 time')
plt.ylabel('quarter-mile time')
plt.title('quarter-mile time vs 0-60 time')
plt.show()
```
